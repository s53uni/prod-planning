### 부품 번호별 
전처리 데이터 저장하기

```python
for part in data["Part Number"].unique() :
    data_part = data[(data["Part Number"]=='{}'.format(part))]
    data_part.reset_index(drop=True, inplace=True)
    
    ### 필요한 컬럼만 추출
    data_part = data_part[['Part Number', 'D일 투입예정 수량(D일계획)', 'D+1일 투입예정 수량(Total)', 'D+2일 투입예정 수량(Total)',
                         'D+3일 투입예정 수량(Total)', 'D+4일 투입예정 수량(Total)', 'D+5일 투입예정 수량', 'D+6일 투입예정 수량',
                         'D+7일 투입예정 수량', 'D+8일 투입예정 수량', 'D+9일 투입예정 수량', 'D+10일 투입예정 수량',
                         'D+11일 투입예정 수량','D+12일 투입예정 수량','CRET_TIME']].copy()
    
    ### 데이터 수집 시간(CRET_TIME) 시간형으로 변환
    data_part['CRET_TIME'] = pd.to_datetime(data_part['CRET_TIME'], format="%Y%m%d%H%M")
    
    ### 하루의 마지막 발주로그만 남기기
    data_part = data_part.groupby(by=[data_part['CRET_TIME'].dt.year,
                                      data_part['CRET_TIME'].dt.month,
                                      data_part['CRET_TIME'].dt.day]).last()
    data_part.reset_index(drop=True, inplace=True)
    
    ### 파트별 데이터 저장
    save_path = './data_new/01_전처리후_파트별_데이터/part{}_data.csv'.format(part[5:])
    data_part.to_csv(save_path, index=False, encoding='utf-8-sig')
```

### 부품 번호별 전처리 데이터 합치기

```python
### 파트별 데이터 합치기
data = pd.read_csv("./data_new/01_전처리후_파트별_데이터/part0_data.csv")

for i in range(1, 117, 1) :
    data_concat = pd.read_csv("./data_new/01_전처리후_파트별_데이터/part{}_data.csv".format(i))
    data = pd.concat([data, data_concat], ignore_index=True)

### 전체 전처리 데이터 저장
save_path = './data_new/02_전처리후_전체_데이터/all_data.csv'
data.to_csv(save_path, index=False, encoding='utf-8-sig')
```

### 1. 부품 번호별 당일 투입예정수량 평균 시각화 

```python
### 전체 전처리 데이터 불러오기
data = pd.read_csv("./data_new/02_전처리후_전체_데이터/all_data.csv")
data
```

```python
### 부품 번호와 당일 발주수량 데이터만 추출
data_plot = data[['Part Number', 'D일 투입예정 수량(D일계획)']].copy()
data_plot['Part Number'] = data_plot['Part Number'].str.slice(5).astype(int)
```

```python
### 데이터 그룹화
data_group = data_plot.groupby(['Part Number']).mean()
data_group.reset_index(inplace=True)

data_group
```

```python
### 부품번호별 당일 투입예정수량 평균 시각화
plt.figure(figsize=(12,4))
plt.bar(data_group['Part Number'], data_group['D일 투입예정 수량(D일계획)'])

plt.title("부품 번호별 당일 투입예정수량 평균")
plt.xlabel('Part Number')
plt.ylabel('당일 투입예정수량')

plt.savefig('./savefig/부품번호별_당일_투입예정수량_평균.png')
```

각 부품마다 평균적으로 발주하는 수량이 다른 것을 확인함<br>
부품들은 비식별화 되어 있어 연관성을 확인하기 어려움

### 2. 부품 번호 및 기준 일자별 당일 투입예정수량 시각화

```python
### 데이터 수집 시간 맞추기
# - 수집시간을 맞추기 위해 모든 수집시간이 다 있는 part0 수집시간 가져옴
data = pd.read_csv("./data_new/01_전처리후_파트별_데이터/part0_data.csv")
date = data['CRET_TIME'].str.slice(0,10).unique()

df = pd.DataFrame(columns=['CRET_TIME'])

df['CRET_TIME'] = date
```

```python
for i in range(0, 117, 1) :
    data = pd.read_csv("./data_new/01_전처리후_파트별_데이터/part{}_data.csv".format(i))
    data = data[['CRET_TIME', 'D일 투입예정 수량(D일계획)']]
    
    data['CRET_TIME'] = data['CRET_TIME'].str.slice(0,10)
    
    df = pd.DataFrame(columns=['CRET_TIME'])
    df['CRET_TIME'] = date
    
    ### 'CRET_TIME'을 기준으로 데이터를 병합
    merged_df = pd.merge(left=df, right=data, how='left')
    merged_df['D일 투입예정 수량(D일계획)'] = merged_df['D일 투입예정 수량(D일계획)'].fillna(0)
    
    ### 파트별 날짜고정 데이터 저장
    save_path = './data_new/03_전처리후_파트별_데이터_날짜고정/part{}_alltime_data.csv'.format(i)
    merged_df.to_csv(save_path, index=False, encoding='utf-8-sig')
    
    ### 파트별 데이터 수집 시간별 D일 계획 시각화
    plt.figure(figsize=(8,10))
    plt.barh(merged_df['CRET_TIME'], merged_df['D일 투입예정 수량(D일계획)'])

    plt.title("Part {} 기준 일자별 당일 투입예정수량".format(i))
    plt.xlabel('당일 투입예정수량')
    plt.ylabel('기준 일자')
    plt.savefig('./savefig/수집시간별계획/part{}_기준일자별_당일투입예정수량.png'.format(i), bbox_inches = 'tight')

    plt.close()
```

시각화 결과, 아예 발주가 없거나 하루 이틀만 발주되는 등의 긴급 발주 부품이 있는 것으로 확인됨<br>
따라서, 긴급 발주 부품을 제외한 일관성 있게 발주되는 부품을 학습하기로 결정함<br>

### 3. 부품 번호별 발주없는 날 카운트

```python
### 부품 번호별 발주없는 날 카운트 및 데이터프레임으로 저장
cnt_list = []
for i in range(0, 117, 1) :
    ### 파트별 날짜고정 데이터 불러오기
    data = pd.read_csv("./data_new/03_전처리후_파트별_데이터_날짜고정/part{}_alltime_data.csv".format(i))
    
    cnt_list.append([i, data[data['D일 투입예정 수량(D일계획)']==0].iloc[:, 1].count()])
count_df = pd.DataFrame(cnt_list, columns=['part_number', '0_cnt'])

save_path = './data_new/부품번호별_발주없는날_카운트.csv'
count_df.to_csv(save_path, index=False, encoding='utf-8-sig')
```

```python
### 데이터프레임 오름차순 정렬해서 확인
count_df.sort_values(by="0_cnt", ascending = True)
```

## 변수별 상관관계 확인

```python
### 예시로 부품 94에 대해서 진행
data_94 = pd.read_csv("./data_new/01_전처리후_파트별_데이터/part94_data.csv")
```

```python
### 변수별 상관관계 파악을 위한 데이터 재배열
corr_data_94 = pd.concat([
    data_94.iloc[12:, :]['D일 투입예정 수량(D일계획)'].reset_index(drop=True),
    data_94.iloc[11:-1, :]['D+1일 투입예정 수량(Total)'].reset_index(drop=True),
    data_94.iloc[10:-2, :]['D+2일 투입예정 수량(Total)'].reset_index(drop=True),
    data_94.iloc[9:-3, :]['D+3일 투입예정 수량(Total)'].reset_index(drop=True),
    data_94.iloc[8:-4, :]['D+4일 투입예정 수량(Total)'].reset_index(drop=True),
    data_94.iloc[7:-5, :]['D+5일 투입예정 수량'].reset_index(drop=True),
    data_94.iloc[6:-6, :]['D+6일 투입예정 수량'].reset_index(drop=True),
    data_94.iloc[5:-7, :]['D+7일 투입예정 수량'].reset_index(drop=True),
    data_94.iloc[4:-8, :]['D+8일 투입예정 수량'].reset_index(drop=True),
    data_94.iloc[3:-9, :]['D+9일 투입예정 수량'].reset_index(drop=True),
    data_94.iloc[2:-10, :]['D+10일 투입예정 수량'].reset_index(drop=True),
    data_94.iloc[1:-11, :]['D+11일 투입예정 수량'].reset_index(drop=True),
    data_94.iloc[:-12, :]['D+12일 투입예정 수량'].reset_index(drop=True)
    ],
    axis=1)
corr_data_94.head()
```

```python
### 상관관계 확인
corr_data_94.corr()
```

현재 목표는 D+1~7 발주 수량을 예측하는 것<br>
당일 발주 수량과 상관관계가 낮아보이는 변수를 확인함<br>
따라서 3일간의 발주 계획(ex.D+1일 계획 + D+2일 계획 + D+3일 계획)을 사용하여 오차를 줄여보고자 함
